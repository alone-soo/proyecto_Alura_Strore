### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.sample(10)
```

```python
tienda2['Precio']
```

#1. Análisis de facturación



```python
ingresot_t1 = tienda['Precio'].sum()
ingresot_t2 = tienda2['Precio'].sum()
ingresot_t3 = tienda3['Precio'].sum()
ingresot_t4 = tienda4['Precio'].sum()
ingreso_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingresot_t1, ingresot_t2, ingresot_t3, ingresot_t4]
})
ingreso_totales['Ingreso Total'] = ingreso_totales['Ingreso Total'].astype(int)
ingreso_totales.plot(kind = 'barh', x = 'Tienda', y = 'Ingreso Total', figsize = (10, 5), title = 'Ingreso Total')
```

# 3. Calificación promedio de la tienda


```python
calificacion_t1 = tienda['Calificación'].mean(skipna= True)
calificacion_t2 = tienda2['Calificación'].mean(skipna= True)
calificacion_t3 = tienda3['Calificación'].mean(skipna= True)
calificacion_t4 = tienda4['Calificación'].mean(skipna=True)
```

```python
calificaciones_tt = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_t1, calificacion_t2, calificacion_t3, calificacion_t4]
})
calificaciones_tt['Calificación Promedio'] = calificaciones_tt['Calificación Promedio'].round(2)
calificaciones_tt.plot(kind = 'line', x = 'Tienda', y = 'Calificación Promedio', figsize = (10, 5), title = 'Calificación', grid= True)
```
